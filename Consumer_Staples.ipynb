```
import pandas as pd 
import numpy as np
import math 
import scipy 
import requests
```

```
import pandas.plotting
import operator
import matplotlib.pyplot as plt
import scipy.stats
import os
import sys
```

```
conStap = pd.read_csv('Con_Staple_Stocks.csv')
```

```
conStap.describe()
```

```
float_prct = operator.truediv(conStap['Float'],conStap['Shares Out'])
```

```
cum_ret = operator.add(conStap['1Y Return'],conStap['3Y Return'])
```

```
cum_rets = operator.add(cum_ret,conStap['5Y Return'])
```

```
diff_52 = operator.sub(conStap['52W High'],conStap['52W Low'])
```

```
rsi_diff = operator.sub(conStap['14D Rel Str'], conStap['50D Rel Str'])
```

```
conStap['Float_Prct']=(float_prct)
conStap['Cummulative_Returns']=(cum_rets)
conStap['Diff_52_Wk']=(diff_52)
conStap['RSI_Diff']=(rsi_diff)
```

```
conStap.describe()
```

```
overbought = conStap.loc[(conStap['14D Rel Str'])>.70]
```

```
oversold = conStap.loc[(conStap['14D Rel Str'])<.55]
```
