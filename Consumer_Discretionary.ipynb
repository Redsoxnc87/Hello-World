```
import pandas.plotting
import scipy.stats
import matplotlib.pyplot as plt
import operator
import os
import sys
```

```
conDis = pd.read_csv('Con_Discretionary_Sctor_Stocks.csv')
```

```
conDis.describe()
```

```
float_prct = operator.truediv(conDis['Float'],conDis['Shares Out'])
```

```
cum_yrRet = operator.add(conDis['1Y Return'],conDis['3Y Return'])
```

```
cum_yrRets = operator.add(cum_yrRet,conDis['5Y Return'])
```

```
diff_52 = operator.sub(conDis['52W High'],conDis['52W Low'])
```

```
conDis['Prct_Float']=(float_prct)
conDis['Cummulative_Returns']=(cum_yrRets)
conDis['Diff_52_HL']=(diff_52)
```

```
conDis.describe()
```

```
rsi_diff = operator.sub(conDis['14D Rel Str'],conDis['50D Rel Str'])
```

```
conDis['Diff_RSis']=(rsi_diff)
```
