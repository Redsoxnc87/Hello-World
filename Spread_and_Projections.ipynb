```
sprd = pd.read_csv('lPE_Hspread.csv')
```

```
qr = operator.mul(sprd['Sales(q)'],4)
sprd['Projected']=(qr)
```

```
CE = operator.sub(sprd['Projected'],sprd['Sales(a)'])
sprd['Crona_Eff']=(CE)
```

```
Gproject = sprd.loc[(sprd['Crona_Eff'])>300000000]
```
