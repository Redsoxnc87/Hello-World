```
import pandas.plotting
import operator
import matplotlib.pyplot as plt
%matplotlib inline
```

```
air = pd.read_csv('Alines_Comp.csv')
```

```
air.head()
```

```
RM = operator.mul(air['Market Capitalization, $K'],1000)
RS = operator.mul(air['Shares Outstanding, K'],1000)
```

```
air['Real_MCAP']=(RM)
air['Real_Shares']=(RS)
```

```
IS = operator.truediv(air['Annual Net Income'],air['Real_Shares'])
SS = operator.truediv(air['Annual Sales'],air['Real_Shares'])
SM = operator.truediv(air['Annual Sales'], air['Real_MCAP'])
```

```
air['Sales_MCAP']=(SM)
air['IN_Shares']=(IS)
air['Sales_Shares']=(SS)
```
